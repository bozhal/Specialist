```python
# 1.1 Создать массив 10 x 10 со случайными значениями от 1 до 100(int64)
arr = np.random.randint(1, 100, (10,10), dtype=np.int64)
print(arr, '\n')

# 1.2 найти среднее значение всех элементов массива np.mean()
arr_mean_all = np.mean(arr)
print(arr_mean_all)

# 1.3 найти среднее значение всех элементов для каждой строки   np.mean()
arr_mean = np.mean(arr, axis = 0)

# и добавить как новый столбец (# 1-я ось - это столбцы 2х мерном массиве)
arr_mean = arr_mean.reshape(10,1); arr_mean

new_arr = np.append(arr, arr_mean, axis=1) # 1-я ось - это столбцы 2х мерном массиве
inf = new_arr.shape
print(new_arr, '\n', inf)

arr_mean_all2 = np.mean(new_arr) # ? Не считает новый столбец!!!
print(arr_mean_all2)

# 1.4 найти сумму значение всех элементов для каждого столбца   np.sum()
arr_sum = np.sum(new_arr, axis=1)
inf2 = arr_sum.shape
print(arr_sum, '\n', inf2)

# и добавить как новую строку, учитывая результат 1.3
arr_sum = arr_sum.reshape(1,-1)
print(arr_sum)
new_arr_sum = np.append(arr, arr_sum, axis=0)
print(new_arr_sum)
```
